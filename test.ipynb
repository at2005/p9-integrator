```python
import rebound
import numpy as np
```

```python
sim = rebound.Simulation()
sim.G = 4*np.pi ** 2
sim.units = ["AU", "yr", "Msun"]
sim.add("Sun")
sim.add("Mercury")
sim.add("Venus")
sim.add("Earth")
sim.add("Mars")
sim.add("Jupiter")
sim.add("Saturn")
sim.add("Uranus")
sim.add("Neptune")
sim.add("Pluto")
sim.add("Eris")
sim.add("Sedna")
sim.move_to_com()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
fig = rebound.OrbitPlot(sim, orbit_style="trail", unitlabel="AU")
```

```python
# numeric integration forward


N_integration = 100000
# pos_earth = np.zeros((N_integration,2))
# pos_jupiter = np.zeros((N_integration, 2))
# pos_comet = np.zeros((N_integration, 2))
pos_sedna = np.zeros((N_integration, 2))
pos_eris = np.zeros((N_integration, 2))
pos_pluto = np.zeros((N_integration, 2))
pos_neptune = np.zeros((N_integration, 2))

timespan = np.linspace(0, -1000, N_integration)


```

```python
sedna = sim.particles[-1]
eris = sim.particles[-2]
pluto = sim.particles[-3]
neptune = sim.particles[-4]

# sedna

for i, time in enumerate(timespan):
    sim.integrate(time)
    pos_sedna[i] = np.array([sedna.x, sedna.y])
    pos_eris[i] = np.array([eris.x, eris.y])
    pos_pluto[i] = np.array([pluto.x, pluto.y])
    pos_neptune[i] = np.array([neptune.x, neptune.y])
    # pos_jupiter[i] = np.array([jupiter.x, jupiter.y])
    # pos_sedna[i] = np.array([comet.x, comet.y])
```

```python
fig, ax = plt.subplots()
# ax.plot(pos_earth[:, 0], pos_earth[:, 1])
# ax.plot(pos_jupiter[:, 0], pos_jupiter[:, 1])
# ax.plot(pos_comet[:, 0], pos_comet[:, 1])

# ax.plot(pos_sedna[:, 0], pos_sedna[:, 1])
ax.plot(pos_eris[:, 0], pos_eris[:, 1])
ax.plot(pos_pluto[:, 0], pos_pluto[:, 1])
ax.plot(pos_neptune[:, 0], pos_neptune[:, 1])
plt.show()
```
